### Логарифмический масштаб

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
masses = [0.055274, 0.815, 1.0, 0.107, 317.8, 95.0, 14.6, 17.147]
```

```python
# без логарифмического масштаба
plt.plot(planets, masses)
plt.show()
```

```python
# логарифмический масштаб
plt.plot(planets, masses)
plt.yscale(value='log')
plt.show()
```

```python
planet_info = pd.Series(masses, index=planets)
planet_info
```

```python
planet_info.plot()
plt.show()
```

```python
# столбчатая диаграмма
planet_info.plot('bar')
plt.show()
```

```python
planet_info.plot('bar', logy = True)
plt.show()
```

```python
planet_info.plot(kind='barh', logx = True)
plt.show()
```

### Форматирование текста для графика

```python
x = np.linspace(-5, 5, 101)
x
```

```python
x2 = x**2
x2
```

```python
plt.plot(x, x2)
plt.show()
```

```python
# добавим название, изменим размер текста, насыщенность текста, укажем цвет из таблицы HTML цветов
plt.plot(x, x2)
plt.title('График f(x)', fontsize=16, fontweight = 'bold', color='DarkRed')
plt.show()
```

```python
# изменим настройки для большей четкости графика
%config InlineBackend.figure_format = 'svg'
```

```python
plt.plot(x, x2)
plt.title('График f(x)', loc='left')
plt.show()
```

Добавление названий осей

```python
plt.plot(x, x2)
plt.title('График')
lable_dict = {'fontsize': '14'}
plt.xlabel('x', fontdict=lable_dict)
plt.ylabel('f(x)', fontdict=lable_dict)
plt.show()
```

```python
x3 = x**3
x3
```

```python
plt.plot(x, x2)
plt.plot(x, x3)
plt.title('График')
lable_dict = {'fontsize': '14'}
plt.xlabel('x', fontdict=lable_dict)
plt.ylabel('f(x)', fontdict=lable_dict)
plt.show()
```

### Ограничение осей

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2)
plt.plot(x, x3)
plt.title('График')
lable_dict = {'fontsize': '14'}
plt.xlabel('x', fontdict=lable_dict)
plt.ylabel('f(x)', fontdict=lable_dict)
plt.show()
```

### Добавление легенды

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2)
plt.plot(x, x3)
plt.title('График')
lable_dict = {'fontsize': '14'}
plt.xlabel('x', fontdict=lable_dict)
plt.ylabel('f(x)', fontdict=lable_dict)
plt.legend(labels=['x2', 'x3'])
plt.show()
```

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')
plt.title('График')
lable_dict = {'fontsize': '14'}
plt.xlabel('x', fontdict=lable_dict)
plt.ylabel('f(x)', fontdict=lable_dict)
legent_dict = {'size': 10}
plt.legend(loc='lower right', prop=legent_dict)
plt.show()
```

```python
# изменяем цвет легенды
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')
plt.title('График')
lable_dict = {'fontsize': '14'}
plt.xlabel('x', fontdict=lable_dict)
plt.ylabel('f(x)', fontdict=lable_dict)

legent_dict = {'size': 10}
legend = plt.legend(loc='lower right', prop=legent_dict, frameon=False)
plt.setp(legend.get_texts(), color='grey')

plt.show()
```

### Оформление линии и сетки графика

```python
# линия графика
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2, label='x**2', color='#3bb08f', linestyle='dashed')
plt.plot(x, x3, label='x**3', color='#e97451', linestyle='dotted')

plt.title('График')
plt.xlabel('x', fontdict=lable_dict)
plt.ylabel('f(x)', fontdict=lable_dict)

legend = plt.legend(loc='lower right', prop=legent_dict, frameon=False)
plt.setp(legend.get_texts(), color='grey')

plt.show()
```

```python
# сетка графика
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('График')
plt.xlabel('x', fontdict=lable_dict)
plt.ylabel('f(x)', fontdict=lable_dict)

legend = plt.legend(loc='lower right', prop=legent_dict, frameon=False)
plt.setp(legend.get_texts(), color='grey')

plt.grid(axis='y', color='lightgrey')
plt.show()
```

### Расположение axes внутри объекта figure

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8.25, 3.5)
plt.subplots_adjust(wspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 6)
plt.subplots_adjust(hspace=0.25)
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title("1")
ax1.set_xlim([-3,3])
ax2.plot(x, x3)
ax2.set_title("2")
ax2.set_ylim([-25, 25])
ax3.plot(x, y)
ax3.set_title("3")
ax4.plot(x, z)
ax4.set_title("4")

fig.set_size_inches(8, 6)
plt.subplots_adjust(hspace=0.3, wspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax2.plot(x, x3)
ax2.set_ylim([-25, 25])
ax3.plot(x, y)
ax4.plot(x, z)

fig.set_size_inches(8, 6)
plt.subplots_adjust(hspace=0.3, wspace=0.3)
```
